```python

import os
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
from sklearn.metrics import f1_score, confusion_matrix
from torch.utils.data import Dataset, DataLoader
import albumentations as A
from albumentations.pytorch import ToTensorV2
import torch
import torch.nn as nn
import segmentation_models_pytorch as smp

# Set base path
BASE_PATH = '/content/drive/MyDrive/Colab Notebooks/Dataset_BUSI_with_GT'
TUMOR_TYPES = ['benign', 'malignant', 'normal']

# Load image paths and labels
image_paths, mask_paths, tumor_labels = [], [], []
for tumor in TUMOR_TYPES:
    folder_path = os.path.join(BASE_PATH, tumor)
    if not os.path.exists(folder_path):
        print(f"Missing folder: {folder_path}")
        continue
    files = os.listdir(folder_path)
    image_files = [f for f in files if f.endswith('.png') and '_mask' not in f]

    for img_file in image_files:
        mask_file = img_file.replace('.png', '_mask.png')
        img_path = os.path.join(folder_path, img_file)
        mask_path = os.path.join(folder_path, mask_file)
        if os.path.exists(img_path) and os.path.exists(mask_path):
            image_paths.append(img_path)
            mask_paths.append(mask_path)
            tumor_labels.append(tumor)
        else:
            print(f"Missing: {img_path} or {mask_path}")

# Create DataFrame
df = pd.DataFrame({
    'image_path': image_paths,
    'mask_path': mask_paths,
    'tumor_type': tumor_labels
})

# Visualize distribution
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='tumor_type', palette='viridis')
plt.title('Tumor Type Distribution')
plt.show()

# Balance classes
df_benign = df[df.tumor_type == 'benign']
df_malignant = resample(df[df.tumor_type == 'malignant'], replace=True, n_samples=len(df_benign), random_state=42)
df_normal = resample(df[df.tumor_type == 'normal'], replace=True, n_samples=len(df_benign), random_state=42)
df = pd.concat([df_benign, df_malignant, df_normal]).sample(frac=1, random_state=42).reset_index(drop=True)

# Visualize balanced distribution
sns.countplot(data=df, x='tumor_type', palette='viridis')
plt.title("Balanced Tumor Type Distribution")
plt.show()

# Dataset class
class BreastUltrasoundDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform
        self.label_map = {'normal': 0, 'benign': 1, 'malignant': 2}

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image = cv2.imread(row.image_path, cv2.IMREAD_GRAYSCALE)
        mask = cv2.imread(row.mask_path, cv2.IMREAD_GRAYSCALE)
        mask = (mask > 0).astype(np.uint8)
        label = self.label_map[row.tumor_type]
        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']
            mask = augmented['mask'].unsqueeze(0)
        return image, mask, label, row.image_path

# Enhanced transforms
train_transform = A.Compose([
    A.Resize(256, 256),
    A.HorizontalFlip(p=0.5),
    A.Rotate(limit=30, p=0.5),
    A.RandomBrightnessContrast(p=0.3),
    A.GaussNoise(p=0.3),
    A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=45, p=0.5),
    A.Normalize(mean=0.5, std=0.5),
    ToTensorV2()
])
val_transform = A.Compose([
    A.Resize(256, 256),
    A.Normalize(mean=0.5, std=0.5),
    ToTensorV2()
])

# Split dataset
train_df, temp_df = train_test_split(df, test_size=0.3, stratify=df.tumor_type, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df.tumor_type, random_state=42)
train_loader = DataLoader(BreastUltrasoundDataset(train_df, train_transform), batch_size=16, shuffle=True)  # Larger batch size for GPU
val_loader = DataLoader(BreastUltrasoundDataset(val_df, val_transform), batch_size=16, shuffle=False)
test_loader = DataLoader(BreastUltrasoundDataset(test_df, val_transform), batch_size=16, shuffle=False)

# Model
class MultiTaskModel(nn.Module):
    def __init__(self, backbone='resnet34', num_classes=3):
        super().__init__()
        self.backbone = smp.UnetPlusPlus(encoder_name=backbone, encoder_weights='imagenet', in_channels=1, classes=1, activation='sigmoid')
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Linear(512, num_classes)

    def forward(self, x):
        seg_output = self.backbone(x)
        enc_feat = self.backbone.encoder(x)[-1]
        cls_output = self.pool(enc_feat).view(enc_feat.size(0), -1)
        cls_output = self.fc(cls_output)
        return seg_output, cls_output

# Loss functions
class DiceLoss(nn.Module):
    def forward(self, pred, target, smooth=1):
        pred, target = pred.contiguous().view(-1), target.contiguous().view(-1)
        intersection = (pred * target).sum()
        return 1 - ((2. * intersection + smooth) / (pred.sum() + target.sum() + smooth))

class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        ce_loss = nn.CrossEntropyLoss(reduction='none')(inputs, targets)
        pt = torch.exp(-ce_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * ce_loss
        return focal_loss.mean()

def combined_loss(seg_pred, seg_target, cls_pred, cls_target, w_seg=0.7, w_cls=0.3):
    dice = DiceLoss()
    seg_loss = 0.5 * dice(seg_pred, seg_target) + 0.5 * nn.BCEWithLogitsLoss()(seg_pred, seg_target)
    cls_loss = FocalLoss()(cls_pred, cls_target)
    return w_seg * seg_loss + w_cls * cls_loss

# Refine predictions with threshold
def refine_predictions(seg_pred, cls_pred, normal_threshold=0.7):
    normal_probs = torch.softmax(cls_pred, dim=1)[:, 0].unsqueeze(1).unsqueeze(2).unsqueeze(3)
    normal_mask = (normal_probs > normal_threshold).float()
    return seg_pred * (1 - normal_mask)

# Training
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = MultiTaskModel().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)  # Lower learning rate for fine-tuning

history = {'loss': [], 'val_loss': [], 'accuracy': [], 'val_accuracy': []}

for epoch in range(40):
    model.train()
    total_loss, correct, total = 0, 0, 0
    for images, masks, labels, _ in tqdm(train_loader):
        images, masks, labels = images.to(device), masks.to(device), labels.to(device)
        optimizer.zero_grad()
        seg_pred, cls_pred = model(images)
        seg_pred = refine_predictions(seg_pred, cls_pred, normal_threshold=0.7)
        loss = combined_loss(seg_pred, masks.float(), cls_pred, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        correct += (cls_pred.argmax(1) == labels).sum().item()
        total += labels.size(0)
    train_acc = correct / total
    train_loss = total_loss / len(train_loader)
    history['loss'].append(train_loss)
    history['accuracy'].append(train_acc)

    model.eval()
    val_loss, val_correct, val_total = 0, 0, 0
    with torch.no_grad():
        for images, masks, labels, _ in val_loader:
            images, masks, labels = images.to(device), masks.to(device), labels.to(device)
            seg_pred, cls_pred = model(images)
            seg_pred = refine_predictions(seg_pred, cls_pred, normal_threshold=0.7)
            loss = combined_loss(seg_pred, masks.float(), cls_pred, labels)
            val_loss += loss.item()
            val_correct += (cls_pred.argmax(1) == labels).sum().item()
            val_total += labels.size(0)
    val_acc = val_correct / val_total
    history['val_loss'].append(val_loss / len(val_loader))
    history['val_accuracy'].append(val_acc)

    print(f"Epoch {epoch+1}: Train Loss={train_loss:.4f}, Train Acc={train_acc:.4f}, Val Loss={val_loss/len(val_loader):.4f}, Val Acc={val_acc:.4f}")

# Save model
torch.save(model.state_dict(), 'model_finetuned.pth')

# Evaluation
def dice_score(pred, target, smooth=1):
    pred = (pred > 0.5).float()
    intersection = (pred * target).sum((1,2,3))
    union = pred.sum((1,2,3)) + target.sum((1,2,3))
    return ((2. * intersection + smooth) / (union + smooth)).mean().item()

def evaluate(model, loader):
    model.eval()
    f1s, dscs = [], []
    with torch.no_grad():
        for images, masks, labels, _ in loader:
            images, masks = images.to(device), masks.to(device)
            seg_pred, cls_pred = model(images)
            seg_pred = refine_predictions(seg_pred, cls_pred, normal_threshold=0.7)
            dscs.append(dice_score(seg_pred, masks))
            f1s.append(f1_score(labels.cpu(), cls_pred.argmax(1).cpu(), average='macro'))
    print(f"DSC: {np.mean(dscs):.4f}, F1: {np.mean(f1s):.4f}")

evaluate(model, test_loader)

# Confusion Matrix
def plot_confusion_matrix(model, loader):
    model.eval()
    y_true, y_pred = [], []
    class_names = ['Normal', 'Benign', 'Malignant']
    with torch.no_grad():
        for images, masks, labels, _ in loader:
            images, labels = images.to(device), labels.to(device)
            _, cls_pred = model(images)
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(cls_pred.argmax(1).cpu().numpy())
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

plot_confusion_matrix(model, test_loader)

# Misclassification Analysis
def analyze_misclassifications(model, loader, num_samples=10):
    model.eval()
    count = 0
    misclassified = []
    class_names = ['Normal', 'Benign', 'Malignant']
    with torch.no_grad():
        for images, masks, labels, paths in loader:
            images, labels = images.to(device), labels.to(device)
            seg_pred, cls_pred = model(images)
            seg_pred = refine_predictions(seg_pred, cls_pred, normal_threshold=0.7)
            preds = cls_pred.argmax(1)
            for i in range(images.size(0)):
                if preds[i] != labels[i]:
                    misclassified.append({
                        'image': images[i].squeeze().cpu().numpy(),
                        'true_label': class_names[labels[i].item()],
                        'pred_label': class_names[preds[i].item()],
                        'path': paths[i],
                        'confidence': torch.softmax(cls_pred[i], dim=0).cpu().numpy()
                    })
                    count += 1
                if count >= num_samples:
                    break
            if count >= num_samples:
                break

    for sample in misclassified:
        plt.figure(figsize=(10, 3))
        plt.subplot(1, 2, 1)
        plt.imshow(sample['image'], cmap='gray')
        plt.title(f"True: {sample['true_label']}, Pred: {sample['pred_label']}")
        plt.axis('off')
        plt.subplot(1, 2, 2)
        plt.bar(class_names, sample['confidence'] * 100)
        plt.title("Prediction Probabilities")
        plt.ylabel("Probability (%)")
        plt.show()
        print(f"Image Path: {sample['path']}")

analyze_misclassifications(model, test_loader)

# Save predictions
def save_predictions(model, loader, out_dir='predictions'):
    os.makedirs(out_dir, exist_ok=True)
    model.eval()
    with torch.no_grad():
        for _, (images, masks, labels, paths) in enumerate(loader):
            images = images.to(device)
            seg_pred, cls_pred = model(images)
            seg_pred = refine_predictions(seg_pred, cls_pred, normal_threshold=0.7)
            for j in range(images.size(0)):
                pred_mask = (seg_pred[j].squeeze().cpu().numpy() > 0.5).astype(np.uint8) * 255
                name = os.path.basename(paths[j])
                cv2.imwrite(os.path.join(out_dir, f'pred_{name}'), pred_mask)

save_predictions(model, test_loader)

# Visualize predictions
def visualize_predictions(model, loader, num_samples=5):
    model.eval()
    count = 0
    with torch.no_grad():
        for images, masks, labels, _ in loader:
            images = images.to(device)
            seg_pred, cls_pred = model(images)
            seg_pred = refine_predictions(seg_pred, cls_pred, normal_threshold=0.7)
            for i in range(images.size(0)):
                if count >= num_samples:
                    return
                img = images[i].squeeze().cpu().numpy()
                true_mask = masks[i].squeeze().cpu().numpy()
                pred_mask = (seg_pred[i].squeeze().cpu().numpy() > 0.5).astype(np.uint8)
                plt.figure(figsize=(10, 3))
                plt.subplot(1, 3, 1)
                plt.imshow(img, cmap='gray')
                plt.title("Image")
                plt.axis('off')
                plt.subplot(1, 3, 2)
                plt.imshow(true_mask, cmap='gray')
                plt.title("Ground Truth")
                plt.axis('off')
                plt.subplot(1, 3, 3)
                plt.imshow(pred_mask, cmap='gray')
                plt.title("Prediction")
                plt.axis('off')
                plt.show()
                count += 1

visualize_predictions(model, test_loader)

# Plot loss and accuracy
plt.figure(figsize=(10, 5))
plt.plot(history['loss'], label='Train Loss')
plt.plot(history['val_loss'], label='Validation Loss')
plt.title('Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(history['accuracy'], label='Train Accuracy')
plt.plot(history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
